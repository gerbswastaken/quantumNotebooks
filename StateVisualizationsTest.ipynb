```python
from qiskit import QuantumCircuit
from qiskit import assemble
from qiskit import Aer

from qiskit.visualization import plot_histogram
from qiskit.visualization import plot_bloch_multivector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_state_qsphere

import numpy as np
```

```python
quantumCircuit = QuantumCircuit(3)

for qubit in range(3):
    quantumCircuit.h(qubit)

quantumCircuit.draw()
```

```python
statevectorSimulator = Aer.get_backend('aer_simulator')
quantumCircuit.save_statevector()
qObject = assemble(quantumCircuit)
stateVector = statevectorSimulator.run(qObject).result().get_statevector()

array_to_latex(stateVector, prefix="\\text{Statevector} = ")
```

```python
qCircuitUnitary = QuantumCircuit(2)
qCircuitUnitary.h(0)
qCircuitUnitary.x(1)
qCircuitUnitary.draw()
```

```python
unitarySimulator = Aer.get_backend("aer_simulator")
qCircuitUnitary.save_unitary()
qObject = assemble(qCircuitUnitary)
unitaryMatrix = unitarySimulator.run(qObject).result().get_unitary()

array_to_latex(unitaryMatrix, prefix="\\text{Unitary} = ")
```

```python
qCircuitBellState = QuantumCircuit(2)
qCircuitBellState.x(1)
qCircuitBellState.barrier()
qCircuitBellState.h(0)
qCircuitBellState.barrier()
qCircuitBellState.cx(0,1)
qCircuitBellState.barrier()
qCircuitBellState.draw()
```

```python
simulator = Aer.get_backend('aer_simulator')
qCircuitBellState.save_statevector()
qObject = assemble(qCircuitBellState)
stateVector = simulator.run(qObject).result().get_statevector()

array_to_latex(stateVector, prefix="\\text{Bell State} =")
```

```python
plot_state_qsphere(stateVector)
```
